```python
from transformers import pipeline
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('output_with_sentiment.csv', sep=',')
df['contents']
```

```python
categories = []
classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)

labels = ["Earnings", "Partnerships", "New Products", "Management", "Regulatory"]

# Iterate through each row using a for loop
for index, row in df.iterrows():
    article = row['contents']
    result = classifier(
        article,
        candidate_labels=labels,
        multi_label=False           
    )
    category = result['labels'][0]

    categories.append(category)  # Store the result

df['category'] = categories
```

```python
df.to_csv('output_with_sentiment_class.csv', sep=',')
```
